```python
%% Example 1.sequence diagram
sequenceDiagram
    participant Consumer
    participant LoanBroker
    participant CreditAgencyGateway
    participant LenderGateway
    participant BankQuoteGateway
    participant Bank1
    participant Bank5

    activate Consumer
    Consumer->>LoanBroker: 1: getLoanQuote
    activate LoanBroker
    LoanBroker->>CreditAgencyGateway: 1.1: getLanQuotes With Scores
    activate CreditAgencyGateway
    CreditAgencyGateway-->>LoanBroker: 1.2: Credit Profile
    deactivate CreditAgencyGateway
    LoanBroker->>LenderGateway: 1.3: getLenderList
    activate LenderGateway
    LenderGateway-->>LoanBroker: 1.4: LenderList
    deactivate LenderGateway
    LoanBroker->>BankQuoteGateway: 1.5: getBestQuote
    activate BankQuoteGateway
    BankQuoteGateway->>Bank1: 1.5.1: getBank Quote
    activate Bank1
    Bank1-->>BankQuoteGateway: 1.5.2: Bank Quote
    deactivate Bank1
    BankQuoteGateway->>Bank5: 1.5.3: getBank Quote
    activate Bank5
    Bank5-->>BankQuoteGateway: 1.5.4: Bank Quote
    deactivate Bank5
    BankQuoteGateway-->>LoanBroker: 1.5.5: getBestQuote
    deactivate BankQuoteGateway
    LoanBroker->>LoanBroker: 1.6: formatReport
    LoanBroker-->>Consumer: 1.7: BestQuoteReport
    deactivate LoanBroker
    deactivate Consumer
```

```python
%% Example 2.sequence diagram
sequenceDiagram
    actor User
    participant CarsManager
    participant Car
    participant WashByWaterStrategy
    participant WashBySteamStrategy

    activate User
    User->>CarsManager: Wash(WashTechnique)
    activate CarsManager
    loop for each Car
        CarsManager->>Car: Wash(WashTechnique)
        activate Car
        alt technique == Water
            Car->>WashByWaterStrategy: Wash(Car)
            activate WashByWaterStrategy
            deactivate WashByWaterStrategy
        else
            Car->>WashBySteamStrategy: Wash(Car)
            activate WashBySteamStrategy
            deactivate WashBySteamStrategy
        end
        deactivate Car
    end
    deactivate CarsManager
    deactivate User
```

```python
%% Example 3.sequence diagram
sequenceDiagram
    participant ActorA
    participant Obj3
    participant Obj4
    participant control
    participant Obj5
    participant ActorB

    activate ActorA
    ActorA->>Obj3: read(att1)
    Obj3-->>ActorA: Return(att2,att3)
    deactivate ActorA
    Obj3->>Obj4: write(att2)
    Obj3->>control: call(att3)
    activate control
    control->>Obj5: read(att3')
    Obj5-->>control: Return(att4)
    deactivate control
    Obj4->>control: write(att4)
    Obj5->>ActorB: write(att5)
    Obj5->>ActorB: write(att6)
```
