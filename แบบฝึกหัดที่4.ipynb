```python
%% Example 1.state diagram
stateDiagram
    [*] --> Off
    Off --> SelfTest : turn on / startup
    SelfTest --> Idle
    Idle --> ServicingCustomer : cardInserted
    ServicingCustomer --> Idle : cancel
    Idle --> Maintenance : service
    Maintenance --> Idle : service
    Maintenance --> OutOfService : failure
    OutOfService --> Off : turn off / shutDown
    OutOfService --> Maintenance : service
    SelfTest --> OutOfService : failure
```

```python
%% Example 2.state diagram
stateDiagram
    [*] --> Created
    Created --> Shipping : request shipment
    Shipping --> Delivered
    Delivered --> Closed : confirm arrival
    Created --> Cancelled : cancel order
    Cancelled --> [*]
    Closed --> [*]
```

```python
%% Example 3.state diagram
stateDiagram
    [*] --> ReadingCard
    ReadingCard --> ReadingPin : Card Read Successfully
    ReadingCard --> EjectingCard : Invalid Card
    ReadingPin --> ChoosingTransaction : Getting Valid Pin
    ReadingPin --> EjectingCard : Invalid Pin
    ChoosingTransaction --> PerformingTransaction : Transaction Choosen
    ChoosingTransaction --> EjectingCard : Cancel Transaction
    PerformingTransaction --> ChoosingTransaction : Another Transaction
    PerformingTransaction --> EjectingCard : Finished Transaction
    EjectingCard --> [*]
```
